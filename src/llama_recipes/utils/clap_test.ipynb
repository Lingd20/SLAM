```python
import numpy as np
import librosa
import torch
import laion_clap

model = laion_clap.CLAP_Module(enable_fusion=False)
model.load_ckpt('/root/models/clap/630k-audioset-best.pt')
```

```python
audio_file = [
    '/root/data/AudioCaps/waveforms/val/YrqfQRErjfk8.wav',
]
audio_embed = model.get_audio_embedding_from_filelist(x = audio_file, use_tensor=True)
print(audio_embed.shape)
```

```python
text = ["A large crowd cheers and applauds",
        "An audience screams and gives applause",
        "Continuous applause and cheering",
        "An audience screams and gives applause",
        "An audience cheers",
        "Audience screams applause cheers"] # <- trap 
text_embed = model.get_text_embedding(text, use_tensor=True)
print(text_embed.shape)
```

```python
import torch
import torch.nn.functional as F

similarity = F.cosine_similarity(audio_embed, text_embed)
print("余弦相似度得分:", similarity)  # 使用.item()将结果转换为Python数值
```
