```python
from datasets import load_dataset
from datasets import load_dataset, Dataset

librispeech_test_clean = load_dataset("librispeech_asr", "clean", split="test",cache_dir="/work/van-speech-nlp/cache")
librispeech_test_other = load_dataset("librispeech_asr", "other", split="test",cache_dir="/work/van-speech-nlp/cache")
```

```python


# Define the local save paths
save_path_clean = "/work/van-speech-nlp/librispeech/librispeech_test_clean"
save_path_other = "/work/van-speech-nlp/librispeech/librispeech_test_other"

# Save the datasets to local disk
librispeech_test_clean.save_to_disk(save_path_clean)
librispeech_test_other.save_to_disk(save_path_other)
```

```python
# To load the datasets from disk later
loaded_librispeech_test_clean = Dataset.load_from_disk(save_path_clean)
loaded_librispeech_test_other = Dataset.load_from_disk(save_path_other)

# Verify the datasets
print(loaded_librispeech_test_clean)
print(loaded_librispeech_test_other)
```

```python
import os
import json

# Initialize a counter and a list to collect the data
counter = 0
data = []

# Iterate through all files in the dataset
for example in loaded_librispeech_test_clean:
    file = example['file']
    audio = example['audio']
    text = example['text'].lower()  # Convert text to lowercase
    speaker_id = example['speaker_id']
    chapter_id = example['chapter_id']
    example_id = example['id']
    file_name = example['id'].split('-')[-1]

    new_file = f"/work/van-speech-nlp/librispeech/LibriSpeech/test-clean/{speaker_id}/{chapter_id}/{example_id}.flac"

    # Check if the new file exists
    if os.path.exists(new_file):
        print("file exists")
        example['file'] = new_file 
        # Create a dictionary for JSONL format
        jsonl_entry = {
            "key": example_id,
            "source": new_file,
            "target": text
        }
        data.append(jsonl_entry)
    else:
        print("file does not exist")

    # Increment the counter
    counter += 1
    


# Write the data to a JSONL file
with open('loaded_librispeech_test_clean.jsonl', 'w') as f:
    for entry in data:
        f.write(json.dumps(entry) + '\n')
```
